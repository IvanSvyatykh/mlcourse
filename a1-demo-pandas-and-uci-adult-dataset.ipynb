```python
import pandas as pd
```

```python
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"
data = pd.read_csv(DATA_URL + "adult.data.csv")
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?**

**2. What is the average age (*age* feature) of women?**

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
(data.groupby("native-country").size() / len(data))["Germany"] * 100
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
mean_age = data[data["salary"] != "<=50K"]["age"].mean()
print(f"Средний возраст: {mean_age}")
std_age = data[data["salary"] != "<=50K"]["age"].std()
print(f"Средне квадратичное отклонение: {std_age}")
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
data[(data["salary"] == ">50K") & (data["education"] == "10th")]
```

Так как существуют люди с образованием 10 классов, получающие более 50 тысяч. Следует это противоречит гипотезе о том, что люди имеющие зп более 50 тысяч имеет хотя бы высшее образование.

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
data.groupby("race").describe()
```

```python
data.groupby("sex").describe()
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
data[(data["sex"] == "Male") & (data["marital-status"].str.startswith("Married"))][
    "salary"
].value_counts(normalize=True)
```

```python
data[(data["sex"] == "Male") & ~(data["marital-status"].str.startswith("Married"))][
    "salary"
].value_counts(normalize=True)
```

Среди мужчин которые не женаты процент, зарабатывающих больше 50 тысяч, больше.

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
max_time = data["hours-per-week"].max()
print(f"Максимальное время работы: {max_time}")
count_people_with_max_time = len(data[data["hours-per-week"] == max_time])
print(f"Количество людей с максимальной занятостью: {count_people_with_max_time}")
percente_of_rich = (
    len(data[(data["hours-per-week"] == max_time) & (data["salary"] == ">50K")])
    * 100
    / count_people_with_max_time
)
print(f"Процент богатых людей с максимальной занятостью : {percente_of_rich}")
```
